```python
#have you ever wondered you are a hacker working for FBI? one of the most important task for a hacker is to decyper and encrypt a password! Let's begin from the very simple case, try to design a system to transform a password to something only you understand. Then try to decypher your classmates' work! this is the code for a caeser cypher, but you can also design your own secrect dictonary.
# Cipher key (each number maps to the number at its index)
cipher_key = [3, 8, 4, 9, 2, 0, 5, 1, 7, 6]

def encrypt(message):
    encrypted_message = ""
    for char in message:
        if char.isdigit():
            # Substitute each digit according to the cipher key
            encrypted_message += str(cipher_key[int(char)])
        else:
            encrypted_message += char  # Keep non-digit characters unchanged
    return encrypted_message

def decrypt(message):
    decrypted_message = ""
    for char in message:
        if char.isdigit():
            # Reverse substitution to get the original number
            decrypted_message += str(cipher_key.index(int(char)))
        else:
            decrypted_message += char  # Keep non-digit characters unchanged
    return decrypted_message

# User interface
while True:
    action = input("Do you want to encrypt or decrypt a message? (type 'exit' to quit): ").lower()
    if action == "exit":
        break
    message = input("Enter the message (numbers only): ")

    if action == "encrypt":
        print("Encrypted message:", encrypt(message))
    elif action == "decrypt":
        print("Decrypted message:", decrypt(message))
    else:
        print("Invalid option. Please type 'encrypt', 'decrypt', or 'exit'.")
```
