We know the total number of heads and feet of all animals in the cage, but we do not know exactly how many chickens and how many rabbits there are. Chickens have two legs, and rabbits have four legs. By using the given total number of heads, H, and the total number of feet, F, calculate how many chickens and rabbits are in the cage.

```python
def find_chickens_and_rabbits(heads, feet):
    # Iterate through all possible numbers of chickens
    for chickens in range(heads + 1):
        rabbits = heads - chickens
        total_feet = 2 * chickens + 4 * rabbits
        if total_feet == feet:
            return chickens, rabbits
    # If no matching case is found after iterating through all possibilities, return None
    return None, None

def find_solution():
  # Input the total number of heads and feet
  heads = int(input("Please enter the total number of heads (H): "))
  feet = int(input("Please enter the total number of feet (F): "))

  # Use the enumeration method to find the answer
  chickens, rabbits = find_chickens_and_rabbits(heads, feet)

  # Output the result
  if chickens is not None and rabbits is not None:
      print(f"There are {chickens} chickens and {rabbits} rabbits.")
  else:
      print("No valid solution was found.")
```

```python
find_solution()
```
