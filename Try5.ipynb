```python
# My Pet Monster Project
#need to know if, loop, def
#Have you ever played some kinds of electronic pets game? now imagine you are raising a pet names "fluffy", you want to design it by python, I have already shown part of the code below, 1. read the code and write an text explaination of what these code means.2. Based on current code, add more functions to your pet (any function you like), like fight with others, or give it a backpack to store foods!
# Variables to keep track of the monster's status
monster_name = "Fluffy"
hunger = 5  # Scale of 1-10, where 10 is very hungry
energy = 5  # Scale of 1-10, where 10 is full of energy
happiness = 5  # Scale of 1-10, where 10 is very happy

def status(evolution = 0):
    print(f"\n{monster_name}'s status: Hunger = {hunger}, Energy = {energy}, Happiness = {happiness}")
    if evolution and hunger >= 10 and energy >= 10 and happiness >= 10:
      print(f"\n{monster_name} has successfully evolved.")
      print(f"1.You need to upgrade your code yourself to complete the subsequent evolution design or even battles.")
      print(f"2.You can design or upload a character for your companions.")
      print(f"For more information, please contact your teacher.")
      return False
    else:
      return True

def feed():
    global hunger, happiness
    if hunger < 9:
        hunger += 2  # Feeding decreases hunger
        happiness += 1
        print(f"\nYou fed {monster_name}. Yum!")
    else:
        print(f"\n{monster_name} is not very hungry right now.")
    status()
    return True

def play():
    global happiness, energy, hunger
    if energy > 2:
        if happiness < 9:
          happiness += 2  # Playing increases happiness
        else:
          happiness = 10
        energy -= 2  # Playing decreases energy
        hunger -= 1
        print(f"\nYou played with {monster_name}. Fun!")
    else:
        print(f"\n{monster_name} is too tired to play.")
    status()
    return True

def sleep():
    global energy, happiness, hunger
    if energy < 8:
        energy += 3  # Sleeping increases energy
        hunger -= 1
        happiness -= 1
        print(f"\n{monster_name} is sleeping. Zzz...")
    else:
        print(f"\n{monster_name} is not very tired, {monster_name} do not want to sleep.")
    status()
    return True

```
