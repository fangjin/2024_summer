```python
# 是否引入坐标概念（可能没学）—— 是否去坐标轴
import matplotlib.pyplot as plt
import numpy as np

def draw_square(x): # 添加功能 如color， 圆可以输入不同参数绘制， 以及加入输出其他参数
    side_length = x # x['side_length']
    color='none'
    plt.gca().add_patch(plt.Rectangle((0, 0), side_length, side_length, edgecolor='black', facecolor= color))
    plt.axis('equal')
    plt.show()

def draw_triangle(side_length):
    plt.gca().add_patch(plt.Polygon([[0, 0], [side_length, 0], [side_length/2, np.sqrt(3)/2*side_length]], edgecolor='black', facecolor='none'))
    plt.axis('equal')
    plt.show()

def draw_circle(radius):
    circle = plt.Circle((radius, radius), radius, edgecolor='black', facecolor='none')
    plt.gca().add_patch(circle)
    plt.axis('equal')
    plt.show()

def draw_rhombus(side_length):
    plt.gca().add_patch(plt.Polygon([[0, side_length/2], [side_length/2, 0], [side_length, side_length/2], [side_length/2, side_length]], edgecolor='black', facecolor='none'))
    plt.axis('equal')
    plt.show()

def main():
    shape = input("Please enter the shape you want to draw (square, triangle, circle, rhombus): ")
    if shape == "square":
        side_length = float(input("Please enter the side length of the square: "))
        draw_square(side_length)
    elif shape == "triangle":
        side_length = float(input("Please enter the side length of the triangle: "))
        draw_triangle(side_length)
    elif shape == "circle":
        radius = float(input("Please enter the radius of the circle: "))
        draw_circle(radius)
    elif shape == "rhombus":
        side_length = float(input("Please enter the side length of the rhombus: "))
        draw_rhombus(side_length)
    else:
        print("Sorry, drawing this shape is not currently supported.")

if __name__ == "__main__":
    main()
```
