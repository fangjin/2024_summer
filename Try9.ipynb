```python
#let's simulate a supermarket self-checkout machine
products = {
    '1': {'name': 'Apple', 'price': 0.50},
    '2': {'name': 'Bread', 'price': 1.20},
    '3': {'name': 'Milk', 'price': 0.99}
}

print("Welcome to the Supermarket Checkout Simulator!")
print("Enter product codes for items you wish to purchase. Enter 'done' when finished.")

total_cost = 0

while True:
    code = input("Enter product code: ").upper()
    if code == 'DONE':
        break
    elif code in products:
        item = products[code]
        total_cost += item['price']
        print(f"Added {item['name']} for ${item['price']:.2f}.")
    else:
        print("Invalid product code. Please try again.")

print(f"Total cost: ${total_cost:.2f}")
payment = float(input("Enter payment amount: $"))

if payment >= total_cost:
    change = payment - total_cost
    print(f"Payment accepted. Your change is ${change:.2f}. Thank you for shopping!")
else:
    print("Insufficient funds. Please provide a sufficient amount for payment.")
```
