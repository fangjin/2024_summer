```python
#simulate an ATM machine
#have you wondered how the vending machine, or the ATM machine works? they allows you to input password and draw money, this is actually, very easy to design, now lets' try to write codes for an ATM machine and explore how it works!
account_number = "123456"
pin = "1234"
balance = 1000  # Starting balance

def login():
    user_account = input("Enter your account number: ")
    user_pin = input("Enter your PIN: ")
    if user_account == account_number and user_pin == pin:
        return True
    else:
        return False

def deposit():
    global balance
    amount = float(input("Enter amount to deposit: "))
    balance += amount
    print(f"${amount} deposited successfully.")

def withdraw():
    global balance
    amount = float(input("Enter amount to withdraw: "))
    if amount <= balance:
        balance -= amount
        print(f"${amount} withdrawn successfully.")
    else:
        print("Insufficient balance.")

def check_balance():
    print(f"Your current balance is: ${balance}")

def main():
    if login():
        print("Login successful!")
        while True:
            print("\n1. Deposit")
            print("2. Withdraw")
            print("3. Check Balance")
            print("4. Exit")
            choice = input("Choose an option (1-4): ")

            if choice == "1":
                deposit()
            elif choice == "2":
                withdraw()
            elif choice == "3":
                check_balance()
            elif choice == "4":
                print("Thank you for using the Mini ATM Simulator. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 4.")
    else:
        print("Invalid account number or PIN.")

if __name__ == "__main__":
    main()
```
